```python
import requests
from  bs4 import BeautifulSoup
import pprint
```

```python
#获取网页源代码
def get_html(url):
    try:
        r = requests.get(url, timeout=30)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return "产生异常"

#获取图片链接并下载
def get_image_and_download(html):
    soup = BeautifulSoup(html, 'html.parser')
    img_list = soup.find_all('img')
    for img in img_list:
        if "/uploads/" not in img['src']:
            continue

        img_url = "https://pic.netbian.com" + img['src']
        img_name = img_url.split('/')[-1]
        with open("beautiful/{}".format(img_name), 'wb') as f:
            f.write(requests.get(img_url).content)
            print("图片{}下载成功".format(img_name))
    
```

```python
if __name__ == "__main__":
    
    htmls = ["https://pic.netbian.com/4kmeinv/"]+["https://pic.netbian.com/4kmeinv/index_{}.html".format(i) for i in range(2,10)]
    for html in htmls:
        html = get_html(html)
        get_image_and_download(html)
```

```python
import requests
from  bs4 import BeautifulSoup
import pprint
import fake_useragent
```

```python
def get_headers():
    ua = fake_useragent.UserAgent()
    headers = {
        'User-Agent': ua.random
    }
    return headers
```

```python
get_headers()
```

```python
def get_novel_chapters():

    url = "https://www.xdingdianxsw.com/0/61/"
    herder = get_headers()
    r = requests.get(url=url,headers=herder)
    r.encoding="gbk"
    if r.status_code != 200:
        raise Exception("request error status code:{}".format(r.status_code))
    soup = BeautifulSoup(r.text, "html.parser")
    links = soup.find("div", class_="article_texttable").find_all("a")
    datas = []
    for link in links:
        href = link.get("href")
        if href is None:
            continue
        href = "https://www.xdingdianxsw.com" + href
        title = link.get("title")
        datas.append((href, title))
    return datas
       

    
```

```python
def get_novel_chapter_content(url):

    
    header = get_headers()
    r = requests.get(url=url,headers=header)
    r.encoding="gbk"
    if r.status_code != 200:
        raise Exception("request error status code:{}".format(r.status_code))
    soup = BeautifulSoup(r.text, "html.parser")
    return  soup.find("div", class_="novel_content").get_text()
```

```python
datas = get_novel_chapters()
```

```python
for url,title in datas:
    content = get_novel_chapter_content(url=url)
    with open("{}.txt".format(title), "w", encoding="utf-8") as f:
        f.write(content)
```
